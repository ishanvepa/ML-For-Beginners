```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection

# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

#print the dimensions of the array (num_patients, num_metrics) and metrics of patient 0
print(X.shape)
print(X[0])
```

```python
X = X[:, 2] #extract just the index 2 column 
print(X.shape)

#reshape to a 2d array
X = X.reshape((-1, 1))
print(X.shape)
```

```python
#split data into training set and testing set
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
#in this case, we are using a 1/3 of the data for training, and 2/3 for testing
```

```python
# create and run linear regression
model = linear_model.LinearRegression()

model.fit(X_train, y_train)
```

```python
#test model by predicting based on X testing set
y_pred = model.predict(X_test)
```

```python
#lets plot the model and data
plt.scatter(X_test, y_test, color='black')

plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xlabel('scaled bmis')
plt.ylabel('disease progression')
plt.title('diabetes progression against bmi')

plt.show()
```
